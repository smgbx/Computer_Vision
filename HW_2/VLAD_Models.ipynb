```matlab
% features of training images
train_features = getImagesFeatures('cdvs_thumbnails\', "sift");

% file location of ground-truth pairs
file = load('MPEG.CDVS\fnames_all-mp_fid-nmp_fid-n_mp-n_nmp.mat');

% Ground truth
match_fid = file.mp_fid(1:25,:);
no_match_fid = file.nmp_fid(1:25,:);
```

# VLAD Models

```matlab
% OPTIONS:
%
% k_dim = 24,48
% num_clus = 32,64,96
% feat_gen = sift, dsft
% agg = vlad, fv
```

```matlab
kd = 24;
k = 32;
feat_gen = "sift";
agg = "vlad";

% Get feature model. Train model with training image set from CDVS
[vlad_m, B] = getVladModel(train_features', kd, k);

% Get features of ground truth sets using generated model
matches = getDistances(match_fid, feat_gen, agg, kd, k, vlad_m);
non_matches = getDistances(no_match_fid, feat_gen, agg, kd, k, vlad_m);

% Using less than 100 instances so it doesn't run forever
train_num = 25;
labels = [ones(1, train_num) -ones(1,train_num)]; 
scores = [matches non_matches];

% Generate ROC
vl_roc(labels, scores);
```
