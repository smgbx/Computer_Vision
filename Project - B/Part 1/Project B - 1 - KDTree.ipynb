```python
from pickle import load, dump
import pandas as pd
from sklearn.neighbors import KDTree
#from scipy.spatial import KDTree
```

```python
data = load(open(r'C:\Users\Shelby\Desktop\UMKC\Academics\Fall20\ComputerVision\Project\data.pkl', 'rb'))
X_train, X_test, y_train, y_test = data[0], data[1], data[2], data[3]
```

```python
# Building kd-tree
kdtree = KDTree(X_train)
```

```python
# Random sample
X_train[1]
```

```python
# Getting 128 NN
dist, ind = kdtree.query(X_train[100].reshape(1,-1), 128)
```

```python
print(dist)
print(ind)
```

```python
def kdtree_query(X):
    all_dist, all_ind = list(), list()
    for img in X:
        dist, ind = kdtree.query(img.reshape(1,-1), 128)
        all_dist.append(dist), all_ind.append(ind)
    return all_dist, all_ind
```

```python
train_dist, train_ind = kdtree_query(X_train)
test_dist, test_ind = kdtree_query(X_test)
```

```python
data = [train_dist, train_ind, test_dist, test_ind]
dump(data, open(r'C:\Users\Shelby\Desktop\UMKC\Academics\Fall20\ComputerVision\Project\kdtree_results','wb'))
```
